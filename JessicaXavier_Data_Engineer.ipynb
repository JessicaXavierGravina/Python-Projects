##__Problem 1: Polygon Optimisation__

Consider the given set of coordinates: <br>
`points = [(1, 1), (1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (3, 3), (3,2), (2,2), (2,1), (1,1)]` <br>
These coordinates outline the following polygon:

###a) Identify the redundant points that are unnecessary for constructing the polygon above and provide a rationale for their exclusion.

The redundant points are:

1. (1,2): It falls on the same horizontal line as (1,3) and (2,2).
2. (2,2): It aligns vertically with (2,1) and (2,3).
3. (3,3): This point lies on the diagonal formed by (3,2) and (3,4).

Removing these points maintains the rectangle's shape and properties, as they do not add new defining characteristics. Thus, the essential vertices are (1,1), (1,3), (3,4), and (3,1), accurately outlining the rectangle.

###b) Create a Python function that takes a list of adjacent points outlining a polygon and returns a simplified version of this list that still accurately represents the same polygon. Please avoid using external libraries for solving this problem. In your solution, include comments to explain each step of the process. You should test your function, but you don't need to send test cases. __Ensure you adhere to the provided function structure and only add your implementation where it says "WRITE YOUR CODE HERE."__

```python
def problem1(coordinate_list):
    """
    Simplify a list of adjacent points outlining a polygon by removing collinear points.

    Args:
        coordinate_list (List[Tuple[int, int]]): List of (x, y) tuples representing the polygon's vertices.

    Returns:
        List[Tuple[int, int]]: Simplified list of (x, y) tuples.
    """
    optimized_polygon = []

    # Ensure at least 3 points for a polygon
    if len(coordinate_list) < 3:
        return coordinate_list

    # Function to determine if three points are collinear
    def is_collinear(p1, p2, p3):
        return (p2[1] - p1[1]) * (p3[0] - p2[0]) == (p3[1] - p2[1]) * (p2[0] - p1[0])

    # Iterate through the points, skipping the last one since it is checked against the first
    for i in range(len(coordinate_list) - 1):
        p1, p2, p3 = coordinate_list[i], coordinate_list[i + 1], coordinate_list[(i + 2) % len(coordinate_list)]
        # If not collinear, add the point to the optimized polygon
        if not is_collinear(p1, p2, p3):
            optimized_polygon.append(p2)

    # Add the last point if it is not collinear with the first and second points
    if not is_collinear(coordinate_list[-2], coordinate_list[-1], coordinate_list[0]):
        optimized_polygon.append(coordinate_list[-1])

    return optimized_polygon

# Test the function with given points
points = [(1, 1), (1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (3, 3), (3, 2), (2, 2), (2, 1), (1, 1)]
simplified_points = problem1(points)
print("Simplified points:", simplified_points)
```

##__Problem 2: Incorrect Syntax with Brackets__

###A) Write a Python function that checks whether a string composed of opening < ( > and closing brackets < ) > is syntactically correct. Please avoid using external libraries for solving this problem. You should test your function, but you don't need to send test cases. __Ensure you adhere to the provided function structure and only add your implementation where it says "WRITE YOUR CODE HERE."__

```python
def problem2(input_string):
    """
    Check whether a string composed of opening <(> and closing <)> brackets is syntactically correct.

    Args:
        input_string (str): The string to be checked. Ex: '(())'

    Returns:
        bool: True if the syntax is correct, False otherwise.
    """
    open_brackets = 0

    for char in input_string:
        if char == '(':
            open_brackets += 1
        elif char == ')':
            if open_brackets == 0:
                return False
            open_brackets -= 1

    return open_brackets == 0

# Test the function
input_string = "(())"
print(problem2(input_string))  # Output: True

input_string = "(()"
print(problem2(input_string))  # Output: False

input_string = "()()"
print(problem2(input_string))  # Output: True
```
